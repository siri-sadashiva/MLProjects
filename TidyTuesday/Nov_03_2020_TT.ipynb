```python
from pandas import DataFrame
df = DataFrame({'count' : ikea.groupby( [ "category"] )['name'].size()}).reset_index()

df.sort_values(by='count')
```

```python
df.plot(kind = 'bar', x = 'category')
```

Tables & desks have the highest number of products, room dividers have lowest number of products

```python
#Get all the rows for 'Tables & desks' category
tabchair = ikea[ikea['category'] == 'Tables & desks']
```

```python
#Group by name
df = DataFrame({'count' : tabchair.groupby( [ "name"] )['designer'].size()}).reset_index()
df.sort_values(by = 'count', ascending = False)
```

```python
df.head(10).plot(kind = 'bar', x = 'name')
```

```python
#Get all rows where name is 'Bekant'
bekant = ikea[ikea['name'] == "BEKANT"]

df = DataFrame({'count' : bekant.groupby( [ "category", "designer","price"] ).size()}).reset_index().sort_values(by='price')
```

```python
df.head(20)
```

```python
df['designer'] = df['designer'].replace(to_replace = 'Eva Lilja Löwenhielm/IKEA of Sweden', value = 'Eva Lilja Löwenhielm')
df['designer'] = df['designer'].replace(to_replace = 'IKEA of Sweden/Eva Lilja Löwenhielm', value = 'Eva Lilja Löwenhielm')
```
